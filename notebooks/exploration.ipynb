## 1. Index a Repository

```python
# Index a sample repository
result = pipeline.index_repository('https://github.com/psf/requests')
print(f"Indexed: {result['repo_name']}")
print(f"Files: {result['total_files']}")
print(f"Chunks: {result['total_chunks']}")
```

## 2. Ask Questions

```python
# Ask about the codebase
questions = [
    "How does authentication work?",
    "What is the session handling mechanism?",
    "Explain the retry logic"
]

for q in questions:
    print(f"\n🤖 Q: {q}")
    answer = pipeline.ask('requests', q)
    print(f"A: {answer[:500]}...")
```

## 3. Visualize Dependencies

```python
# Generate and visualize architecture
arch = pipeline.generate_architecture('requests')

# Display metrics
metrics = arch['metrics']
print(f"Components: {metrics['total_components']}")
print(f"Dependencies: {metrics['total_dependencies']}")
print(f"Average Degree: {metrics['average_degree']:.2f}")

# Plot centrality
central = metrics['central_components'][:10]
names = [c[0] for c in central]
values = [c[1] for c in central]

plt.figure(figsize=(10, 6))
plt.barh(names, values)
plt.xlabel('Centrality')
plt.title('Most Central Components')
plt.tight_layout()
plt.show()
```

## 4. Bug Detection

```python
# Find potential bugs
bugs = pipeline.find_bugs('requests')

print(f"Found {len(bugs)} potential issues:")
for i, bug in enumerate(bugs[:5]):
    print(f"\n{i+1}. {bug.get('title', 'Unknown')}")
    for detail in bug.get('details', [])[:3]:
        print(f"   - {detail}")
```
